∣x1,x2⟩∣0⟩→∣x1,x2⟩∣x1∨x2⟩∣x1​,x2​⟩∣0⟩→∣x1​,x2​⟩∣x1​∨x2​⟩, where ∨∨ denotes the logical OR.

```python
from qiskit import QuantumCircuit,QuantumRegister,ClassicalRegister
from qiskit.visualization import plot_histogram
```

```python
#Solution 1 Self made using de morgans law NOT(NOT A AND NOT B)== OR
crqt=QuantumCircuit(4)
crqt.x(0)
crqt.x(1)
crqt.ccx(0,1,2)
crqt.x(2)
crqt.cx(2,3)
crqt.x(2)
crqt.ccx(0,1,2)
crqt.x(1)
crqt.x(0)
display(crqt.draw("mpl"))
#Bad because more logic operations and not the most efficient
#Using X gates might interfere with entagled bits
```
